```python
import pandas as pd
data = pd.read_csv("Accommodation.csv")
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print(data.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=data)
plt.title('Accommodation Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='AddressRegion', data=data)
plt.title('Distribution of Accommodation by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
```

```python
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)  # Binomial(n=10, p=0.5)
plt.figure(figsize=(10, 6))
sns.histplot(binomial_data, kde=True, color='skyblue')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
poisson_data = np.random.poisson(lam=5, size=1000)  # Poisson(lam=5)
plt.figure(figsize=(10, 6))
sns.histplot(poisson_data, kde=True, color='salmon')
plt.title('Poisson Distribution (lam=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
normal_data = np.random.normal(loc=0, scale=1, size=1000)  # Normal(mean=0, std=1)
plt.figure(figsize=(10, 6))
sns.histplot(normal_data, kde=True, color='green')
plt.title('Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
